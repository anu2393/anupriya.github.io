```python
# We now create a dictionary,speakerMap, which will map the names to the dialogues.
# As done in the above cells, we split the dialogue at the 1st index with ":"
#We are replacing ",",".","?","''" with empty string and removing the white spaces from the dialogues
# We then convert the dialogues to lowercase
# when the 1st speaker, WILL is searched and not found, it is given an empty value, where his dialogues are added using extend()
# the loop iterates for the entire text


speakerList = []
speakerMap = {}
for line in f:
    speaker = line.split(":")[0]
    dialogue = line.split(":")[1].replace("…","").replace(",","").replace(".","").replace("?","").replace("!","").strip().split(" ")
    for i in range(len(dialogue)):
        dialogue[i] = dialogue[i].lower()
    if speakerMap.get(speaker) == None:
        speakerMap[speaker] = []
        
    speakerMap[speaker].extend(dialogue)  
```

```python
#converting to set to get unique and back to list

for k,v in speakerMap.items():
    speakerMap[k] = list(set(v))
```

```python
#save to file
#joining the values at every lines
# writing and saving to the text file to the key (or the speaker name)

for k,v in speakerMap.items():
    st = "\n".join(v)
    with open(k+'.txt', 'w',encoding="utf-8") as the_file:
        the_file.write(st)
    
```
